```python
import numpy as np
import pandas as pd

import sklearn

import matplotlib

import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape)
type(digits.data)
```

```python
plt.gray()
plt.matshow(digits.images[0])
plt.show()
```

```python
data, target = digits.data, digits.target
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
pca = PCA(n_components=2)
pca_value = pca.fit_transform(data)
x,y = pca_value.transpose()

plt.scatter(x,y)
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

pca = PCA(n_components=2)
pca_value = pca.fit_transform(data)
x,y = pca_value.transpose()

fig, ax = plt.subplots(figsize=(12, 12))

for x0, y0, image in zip(x, y, digits.images):
    ab = AnnotationBbox(OffsetImage(image, zoom=2), (x0, y0), frameon=False)
        
    ax.add_artist(ab)

plt.scatter(x,y)
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

pca = PCA(n_components=2)
pca_value = pca.fit_transform(data)
x,y = pca_value.transpose()

fig, ax = plt.subplots(figsize=(12, 12))

color_select = ['blue', 'green', 'red',
                'cyan', 'magenta', 'yellow',
                '#496F5D', '#4c9f70', '#49596B',
                '#6279B8']
colors = []

for t in digits.target:
    colors.append(color_select[t])

plt.scatter(x, y, c=colors)
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

x, y, z = PCA(n_components=3).fit_transform(data).transpose()

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')

color_select = ['blue', 'green', 'red',
                'cyan', 'magenta', 'yellow',
                '#496F5D', '#4c9f70', '#49596B',
                '#6279B8']
colors = []

for t in digits.target:
    colors.append(color_select[t])

plt.scatter(x, y, z, c=colors)
```
